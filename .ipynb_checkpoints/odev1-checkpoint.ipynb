# Soru 4
*Bayes Teoreminin gerçek hayatta karşımıza çıkan ve kullanılan örneklerinden 5 tane bulalım.*


Bayes Teoremi'nde "P(A\B) = P(B\A)P(A) / P(B) " formülünü kullanarak aşağıdaki örnekler çözülebilir:
#### Örnek1
Bir kişide gluten hassasiyeti varsa, o kişinin çölyak hastası olma ihtimalini bulmak istiyoruz. 

Verilere göre, bir klinikteki hastaların yüzde 10'unun çölyak hastası ise P(A) = 0,10
Klinikteki hastaların yüzde 5'inin gluten hassasiyeti olduğunu göstermektedir. P(B) = 0,05
Klinik bulguları, gluten hassasiyeti olan hastaların yüzde 7'sinin çölyak hastası olduğunu da göstermektedir. Yani, gluten hassasiyeti olan hastanın çölyak hastası olma olasılığı yüzde 7'dir. (B\A) = 0,07

P (A\B) = (0.07 * 0.10) / (0.05) = 0.14
Yani, bir kişi çölyak hastası ise, gluten hassasiyetine sahip olma olasılığı yüzde 14'tür.

#### Örnek2
Bir kanser testinin sonuçlarına göre hasta kişinin kanser olma olasılığını bulmak istiyoruz.  Bu kişilere sonucu pozitif ve negatif çıkacak testler yapılmıştır. Bu test sonucu pozitif çıkan bir hastanın kanser olma olasılığını hesaplamak istiyoruz.

Popülasyonda 100 kişiden birinin kanser olacağına bildiğimizi varsayalım. P(A) = 0.01 
Hastalığın olduğu durumda %80 oranında pozitif. P(B\A) = 0.80
Hastalığın olmadığı durumda %10  
Testin pozitif çıkma olasığı P(B) = 0.01 * 0.80 + 0.99 * 0.10 = 0.107 

P (A\B) = (0.01 * 0.80) / (0.10) = 0.07
Yani, testi pozitif çıkan birinin kanser olma olasılığı yüzde 7'dir.

#### Örnek3
A ve B olan iki ürünün üretimi planlanmaktadır. 
A ürününün planlanan miktarda üretilme olasılığı %70  P(A) =0.70
B ürününün planlanan miktarda üretilme olasılığı %80  P(B)=0.80
A ürünü planlanan miktarda üretilirse, B ürünün planlanan miktarda üretilme olasılığı P(B\A)= %0.1
Her iki ürünün de planlanan miktarda üretilme olasılığına bakarsak:

P (A\B) = (0.1 * 0.70) / (0.80) = 0.08
Yani, her iki ürünün de planlanan miktarda üretilme olasılığı yüzde 8'dir.


#### Örnek4
Bir okul öğrencilerinde COVID-19 virüsü olup olmadığını test etmek istiyor. 
100 öğrenciden 20'sinin virüs taşıdığını varsayalım. P(A) = 0.20
Yapılan testler virüs taşıma durumunda %96 pozitif sonuç vermekte. P(B\A) = 0.96
Yapılan testler virüs taşımama durumunda %2 pozitif sonuç vermekte.  
Testin pozitif çıkma olasığı P(B) = 0.20 * 0.96 + 0.80 * 0.02 = 0.208

P (A\B) = (0.20 * 0.96) / (0.208) = 0.92
Yani, testi pozitif çıkan birinin virüs taşıma olasılığı yüzde 92'dir.


#### Örnek5
Ampül üretrimi yapan üç fabrikaya bakalım. Ampüllerin %20'si A fabrikasında, %50'si B fabrikasında ve %30'u C fabrikasında üretilmektedir.
A fabrikasında üretilen ampullerin %2'si, B fabrikasında üretilen ampullerin %1'i ve C fabrikasında üretilen ampullerin %3'ü arızalıdır.
Piyasada rastgele bir ampul seçiliyor ve arızalı olduğu tespit ediliyor. Bu ampulün B fabrikasında üretilmiş olma olasılığına bakalım:

P(A)=0.20
P(B)=0.50
P(C)=0.30
Arızalı bir ampulün seçilme olasılığı = P(D) 
P(D\A)=0.02
P(D\B)=0.01
P(D\C)=0.03


P(B\D)= P(D\B)P(B)  /  P(D\A)P(A)+P(D\B)P(B)+P(D\C)P(C)

P(B\D)=(0.01*0.50) / (0.02*0.20)+ (0.01*0.50)+ (0.03*0.30)=0.2777
Yani, arızalı ampülün B pabrikasında üretilmiş olma olasılığı 28'dir.
